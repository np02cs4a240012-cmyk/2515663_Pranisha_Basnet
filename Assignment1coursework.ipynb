```python
import pandas as pd
import numpy as np
import pycountry

# Function to standardize country names
def standardize_country(name):
    try:
        return pycountry.countries.lookup(name).name
    except:
        return name

# Clean column names
hdi_2022_df.columns = hdi_2022_df.columns.str.strip()

# Replace special characters with NaN
hdi_2022_df.replace(["–", "-", "..", "NA", "N/A", ""], np.nan, inplace=True)

# Standardize country names using pycountry
if "Country" in hdi_2022_df.columns:
    hdi_2022_df["Country"] = (
        hdi_2022_df["Country"]
        .astype(str)
        .str.strip()
        .apply(standardize_country)
    )

# Convert numeric columns stored as text
for col in hdi_2022_df.columns:
    if hdi_2022_df[col].dtype == "object" and col != "Country":
        hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors="ignore")

# Remove duplicate rows
hdi_2022_df.drop_duplicates(inplace=True)

# Fill missing values in numeric columns using mean
numeric_cols = hdi_2022_df.select_dtypes(include=["int64", "float64"]).columns
hdi_2022_df[numeric_cols] = hdi_2022_df[numeric_cols].fillna(
    hdi_2022_df[numeric_cols].mean()
)

# Drop rows with missing categorical values
hdi_2022_df.dropna(inplace=True)

# Final check
print("Missing values after cleaning")
print(hdi_2022_df.isna().sum())

print("Final dataset shape")
print(hdi_2022_df.shape)

hdi_2022_df.head()
```

```python
import pandas as pd
import numpy as np
import pycountry

# Load filtered Problem 1B dataset
df = pd.read_csv("HDI_problem1B.csv", encoding="latin1")

# Clean column names
df.columns = df.columns.str.strip()

# Find essential columns safely
hdi_col = [col for col in df.columns if col.lower() == "hdi"][0]
year_col = [col for col in df.columns if col.lower() == "year"][0]
country_col = [col for col in df.columns if "country" in col.lower()][0]

# Check missing values in essential columns
print("Missing values in essential columns")
print(df[[hdi_col, year_col, country_col]].isna().sum())

# Replace non numeric symbols with NaN
df.replace(["–", "-", "..", "NA", "N/A", ""], np.nan, inplace=True)

# Standardize country names using pycountry
def standardize_country(name):
    try:
        return pycountry.countries.lookup(name).name
    except:
        return name

df[country_col] = (
    df[country_col]
    .astype(str)
    .str.strip()
    .apply(standardize_country)
)

# Convert numeric columns stored as text
df[hdi_col] = pd.to_numeric(df[hdi_col], errors="coerce")
df[year_col] = pd.to_numeric(df[year_col], errors="coerce")

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Handle missing values
# Drop rows with missing country or year
df.dropna(subset=[country_col, year_col], inplace=True)

# Fill missing HDI values using mean
df[hdi_col] = df[hdi_col].fillna(df[hdi_col].mean())

# Final check
print("\nMissing values after cleaning")
print(df[[hdi_col, year_col, country_col]].isna().sum())

print("\nFinal dataset shape")
print(df.shape)

df.head()
```

3. Outlier Detection:
• Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.
• Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.
• Discuss why the identified countries stand out as outliers.

```python
import pandas as pd

df = pd.read_csv("Human_Development_Index_Dataset.csv", encoding="latin1")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# South Asia country list
south_asia_countries = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

# Create South Asia dataset (ALL YEARS)
hdi_southasia_2020_2022 = df[
    df['country'].isin(south_asia_countries)
].copy()

# Calculate IQR for HDI
Q1_hdi = hdi_southasia_2020_2022['hdi'].quantile(0.25)
Q3_hdi = hdi_southasia_2020_2022['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

# Calculate IQR for GNI per Capita
Q1_gni = hdi_southasia_2020_2022['gross_inc_percap'].quantile(0.25)
Q3_gni = hdi_southasia_2020_2022['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

# Detect outliers using 1.5 × IQR rule
hdi_southasia_2020_2022['outlier'] = (
    (hdi_southasia_2020_2022['hdi'] < Q1_hdi - 1.5 * IQR_hdi) |
    (hdi_southasia_2020_2022['hdi'] > Q3_hdi + 1.5 * IQR_hdi) |
    (hdi_southasia_2020_2022['gross_inc_percap'] < Q1_gni - 1.5 * IQR_gni) |
    (hdi_southasia_2020_2022['gross_inc_percap'] > Q3_gni + 1.5 * IQR_gni)
)

# Split normal vs outliers
normal = hdi_southasia_2020_2022[hdi_southasia_2020_2022['outlier'] == False]
outliers = hdi_southasia_2020_2022[hdi_southasia_2020_2022['outlier'] == True]

# Scatter plot with TWO COLORS
plt.figure(figsize=(8,6))

plt.scatter(
    normal['gross_inc_percap'],
    normal['hdi'],
    color='blue',
    label='Normal Countries'
)

plt.scatter(
    outliers['gross_inc_percap'],
    outliers['hdi'],
    color='red',
    label='Outlier Countries'
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Outlier Detection: GNI per Capita vs HDI (South Asia, All Years)")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create GNI HDI Gap
hdi_southasia_2020_2022['GNI_HDI_Gap'] = (
    hdi_southasia_2020_2022['gross_inc_percap'] -
    hdi_southasia_2020_2022['hdi']
)

# Aggregate by country (average over 2020–2022)
gap_avg = (
    hdi_southasia_2020_2022
    .groupby('country')[['GNI_HDI_Gap']]
    .mean()
    .reset_index()
)

# Top 3 positive gaps
top_positive = gap_avg.sort_values(
    by='GNI_HDI_Gap',
    ascending=False
).head(3)

# Top 3 negative gaps
top_negative = gap_avg.sort_values(
    by='GNI_HDI_Gap',
    ascending=True
).head(3)

print("Top 3 countries with highest positive GNI-HDI gap:")
print(top_positive)

print("\nTop 3 countries with highest negative GNI-HDI gap:")

# Plot top 3 positive gaps
plt.figure(figsize=(8,4))
plt.bar(
    top_positive['country'],
    top_positive['GNI_HDI_Gap'],
    color='orange'
)
plt.title('Top 3 South Asian Countries with Positive GNI-HDI Gap')
plt.ylabel('GNI-HDI Gap')
plt.show()

# Plot top 3 negative gaps
plt.figure(figsize=(8,4))
plt.bar(
    top_negative['country'],
    top_negative['GNI_HDI_Gap'],
    color='purple'
)
plt.title('Top 3 South Asian Countries with Negative GNI-HDI Gap')
plt.ylabel('GNI-HDI Gap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Human_Development_Index_Dataset.csv", encoding="latin1")

# South Asia countries
south_asia_countries = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

# Middle East countries
middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

# Filter South Asia (2020–2022)
hdi_southasia_2020_2022 = df[
    (df['country'].isin(south_asia_countries)) &
    (df['year'].isin([2020, 2021, 2022]))
]

# Filter Middle East (2020–2022)
hdi_middleeast_2020_2022 = df[
    (df['country'].isin(middle_east_countries)) &
    (df['year'].isin([2020, 2021, 2022]))
]

# Metrics to compare
metrics = {
    'gender_development': 'Gender Development Index',
    'life_expectancy': 'Life Expectancy Index',
    'gross_inc_percap': 'GNI per Capita Index'
}

# Calculate regional means
sa_means = hdi_southasia_2020_2022[list(metrics.keys())].mean()
me_means = hdi_middleeast_2020_2022[list(metrics.keys())].mean()

comparison_df = pd.DataFrame({
    'South Asia': sa_means,
    'Middle East': me_means
})

# Plot grouped bar charts
for metric, title in metrics.items():
    comparison_df.loc[metric].plot(kind='bar', figsize=(6,4))
    plt.title(title + ' Comparison (2020–2022)')
    plt.ylabel(title)
    plt.xticks(rotation=0)
    plt.show()
```

5. HDI Disparity:
• Compute the range (max – min) of HDI for each region.
• Compute the coefficient of variation (CV = std/mean) for HDI.
• Identify which region exhibits more variation in HDI.

```python
import pandas as pd

# Compute HDI statistics for South Asia
south_hdi = hdi_southasia_2020_2022['hdi']
south_range = south_hdi.max() - south_hdi.min()
south_cv = south_hdi.std() / south_hdi.mean()

# Compute HDI statistics for Middle East
middle_hdi = hdi_middleeast_2020_2022['hdi']
middle_range = middle_hdi.max() - middle_hdi.min()
middle_cv = middle_hdi.std() / middle_hdi.mean()

# Display results
print("HDI Disparity Measures (2020–2022)")
print("\nSouth Asia:")
print("HDI Range:", round(south_range, 3))
print("HDI Coefficient of Variation:", round(south_cv, 3))

print("\nMiddle East:")
print("HDI Range:", round(middle_range, 3))
print("HDI Coefficient of Variation:", round(middle_cv, 3))
```

7. Outlier Detection:
• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.
• Create scatter plots highlighting outliers in a different color.
• Discuss the significance of these outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to detect outliers using IQR
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

for region_name, df in {
    'South Asia': hdi_southasia_2020_2022,
    'Middle East': hdi_middleeast_2020_2022
}.items():

    # Detect outliers
    hdi_outliers = detect_outliers(df['hdi'])
    gni_outliers = detect_outliers(df['gross_inc_percap'])
    combined_outliers = hdi_outliers | gni_outliers

    print(f"\n{region_name} Outlier Countries:")
    print(df.loc[combined_outliers, ['country', 'hdi', 'gross_inc_percap']])

    # Scatter plot
    plt.figure(figsize=(6,5))
    plt.scatter(df['gross_inc_percap'], df['hdi'], label='Normal')
    plt.scatter(df.loc[combined_outliers, 'gross_inc_percap'],
                df.loc[combined_outliers, 'hdi'],
                label='Outliers')

    plt.xlabel('GNI per Capita')
    plt.ylabel('HDI')
    plt.title(f'{region_name}: HDI vs GNI per Capita (Outliers Highlighted)')
    plt.legend()
    plt.show()
```
